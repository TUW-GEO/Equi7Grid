# How do I start?

This page provides an easy entry point on how to create and use your first `Equi7Grid` object. The Equi7Grid is a grid framework that is a definition for the full globe and organises the continental zones and tiling systems. These tiling systems share the same tiling concept and the same map projection functions, but differ in their projection parameters that are centred on the different continents. Specifically, for the Equi7Grid seven continental equidistant azimuthal projections are used.

Equi7Grid's seven continental tiling schemes are named as following:
- Africa ("AF")
- Antarctica ("AN")
- Asia ("AS")
- Europe ("EU")
- North America ("NA")
- Oceania ("OC")
- South America ("SA").

By default, each continent is a consistent tiling system with three tiling levels by default:
- T1 (100km tile size)
- T3 (300km tile size)
- T6 (600km tile size)
There is the option for other tiling levels to be defined by the user.

Since the Equi7Grid was primilary defined for high-resolution raster data over land surfaces, a speciality of the `Equi7Grid` realisation is that it seperates between land and ocean internally.


## Initialisation

Lets start with how to initialise a standard `Equi7Grid` object. The package offers the user-facing function `get_standard_equi7grid` to create an `Equi7Grid` instance from a given sampling. If you have a certain tile size in mind and you want to know which samplings are compatible with it, you can use the `allowed_samplings` function. In the following example we want to check, what samplings fit into a 600km tile size ("T1").

```python
from equi7grid import allowed_samplings

allowed_samplings(600_000)
```

We decide to pick 1km as our desired sampling and can now pass this onto `get_standard_equi7grid`. 

```python
from equi7grid import get_standard_equi7grid

e7grid = get_standard_equi7grid(1000)
e7grid
```

You just created your first `Equi7Grid` instance! Note that this command sets the sampling of all standard tiling levels ("T1", "T3", and "T6") to 1km. Please checkout the more detailed [grid](https://tuw-geo.github.io/Equi7Grid/latest/guides/grid.html#initialisation) documention to learn how to initialise an `Equi7Grid` object in a different way or to define samplings per tiling level.

The `e7grid` object provides several global functions for retrieving [coordinates](https://tuw-geo.github.io/Equi7Grid/latest/guides/grid.html#geographic-coordinate-transformation) and [tiles](https://tuw-geo.github.io/Equi7Grid/latest/guides/grid.html#tile-queries) across different continents. More details can be found in the respective documentation.

## Tiling system

As you can see in the printed output of the `e7grid` object, each Equi7Grid continent is represented by an `Equi7TilingSystem` object. You can simply access these instances via the given shortcut and print some details:

```python
print(e7grid.AF)  # noqa: T201
```

The tiling system defines its own projection, projection zone, land zone, and overlays the projection extent with the hierarchy of pre-defined regular tilings. Similar to `e7grid`, it provides methods for retrieving [coordinates](https://tuw-geo.github.io/Equi7Grid/latest/guides/tiling_system.html#projection-system-interactions) and [tiles](https://tuw-geo.github.io/Equi7Grid/latest/guides/tiling_system.html#tiles) on a continent level. 

## Tile

The `Equi7Tile` is the last object and smallest unit in the `EquiGrid` system hierarchy. It can be retrieved by executing the tile queries of the `Equi7Grid` and `Equi7TilingSystem` systems. All functions work in the same manner, with a `cover_land` flag to filter for tiles covering land and a generator object as a return value. As an example, you can retrieve all tiles accross all continents intersecting with the given geographical bounding box with `get_tiles_in_geog_bbox`.

```python
e7tiles = list(
    e7grid.get_tiles_in_geog_bbox(bbox=(0, 30, 10, 40), tiling_id="T6", cover_land=True)
)
[e7tile.name for e7tile in e7tiles]
```

An `Equi7Tile` instance represents a raster covering a certain space in an Equi7Grid projection with the sampling defined by the grid/tilings.

```python
e7tile = e7tiles[0]
print(e7tile)  # noqa: T201
```

It provides several methods for interacting with world and pixel system coordinates, its raster boundary and extent, and its topological relation with other geometries. Please checkout the [tile](https://tuw-geo.github.io/Equi7Grid/latest/guides/tile.html#tile) documentation, which explains all `Equi7Tile` properties and methods in detail.
